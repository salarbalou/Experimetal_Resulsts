```python
import numpy as np
import scipy as sc
import plotly.graph_objects as go
from plotly.subplots import make_subplots
pio.templates.default = 'plotly_dark'
import plotly.io as pio
import re
import requests as req
import scipy.optimize as opt
```

```python
regex = re.compile("^([0-9.]+)\s[,]\s([0-9.]+)\s[,]\s([0-9.]+)\s[,]\s([0-9.]+)\s[,]\s([0-9.]+)", re.MULTILINE)
datatext = req.get("https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/9-14-22-ThermalCycling_Best%20Result.txt").text
datalist = regex.findall(datatext)
datat = []
dataT = []
dataa = []
datab = []
datac = []
for i in range (250):
   datat.append(float(datalist[i][0]))
for j in range (250):
   dataT.append(float(datalist[j][1]))
for k in range (250):
   dataa.append(float(datalist[k][2]))
for m in range (250):
   datab.append(float(datalist[m][3]))
t = np.array(datat)
T = np.array(dataT)
a = np.array(dataa)
b = np.array(datab)
#a[np.where(b!=0)[0]] = b[np.where(b!=0)[0]]
a[93:137] = b[93:137]
a[164:210] = b[164:210]
```

```python
fig = make_subplots(rows= 2, cols= 1)
fig.add_trace(go.Scatter(x = t, y = a, name = 'PWM Profile'), row = 1, col = 1)
#fig.add_trace(go.Scatter(x = t, y = b, name = 'PWM Profile'), row = 1, col = 1)
fig.add_trace(go.Scatter(x = np.linspace(0, t[-1], 100), y = np.ones(250)*95, name = 'Setpoint Denaturation'), row = 2, col = 1)
fig.add_trace(go.Scatter(x = np.linspace(0, t[-1], 100), y = np.ones(250)*60, name = 'Setpoint Extension'), row = 2, col = 1)
fig.add_trace(go.Scatter(x = t, y = T, name = 'Temperature Profile', line_color ='red'), row = 2, col = 1)
fig.update_xaxes(title = 'time (s)', title_font = dict(size = 16), range = [0, 250], tickfont = dict(size = 16))
fig.update_yaxes(title = 'PWM of Arduino',title_font = dict(size = 16), row= 1, col = 1, tickfont = dict(size = 16))
fig.update_yaxes(title = 'Temperature (ËšC)',title_font = dict(size = 16), row= 2, col = 1, tickfont = dict(size = 16))
fig.update_layout(height = 1000, width = 1000, legend = dict(title = f'kp = 20, ki = 2.5', font= dict(size = 16)))
```
