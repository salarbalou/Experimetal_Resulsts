```python
import numpy as np
import plotly.io as pio
pio.templates.default  = 'plotly_white'
from plotly.subplots import make_subplots
import plotly.graph_objects as go
import plotly.express as px
import scipy.optimize as scop
```

```python
dataB = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/Blue_CQD_FTIR.txt')
dataC = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/Cyan_CQD_FTIR.txt')
dataG = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/Orange_CQD_FTIR.txt')
dataY = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/Yellow_CQD_FTIR.txt')
dataR = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/Red_CQD_FTIR.txt')

waveB, intentB = dataB[:, 0], dataB[:, 1]
waveC, intentC = dataC[:, 0], dataC[:, 1]
waveG, intentG = dataG[:, 0], dataG[:, 1]
waveY, intentY = dataY[:, 0], dataY[:, 1]
waveR, intentR = dataR[:, 0], dataR[:, 1]

#peaks = np.unique(np.array(wave[np.argmax(intent)], dtype = 'float64'))
#noise = np.random.normal(loc = 1, scale = 10**(-n), size = 7001)
r = 40

figure = make_subplots()

figure.add_trace(go.Scatter(x = waveB, y = intentB+4*r, mode= 'lines', line_color = 'blue'))
figure.add_trace(go.Scatter(x = waveC, y = intentC+3*r, mode= 'lines', line_color = 'cyan'))
figure.add_trace(go.Scatter(x = waveG, y = intentG+2*r, mode= 'lines', line_color = 'green'))
figure.add_trace(go.Scatter(x = waveY, y = intentY+r, mode= 'lines', line_color = 'orange'))
figure.add_trace(go.Scatter(x = waveR, y = intentR, mode= 'lines', line_color = 'red'))

figure.update_xaxes(title = dict(text = 'Wavelength (nm)', font = dict(size = 20)), ticks = 'outside', showgrid = False, showline = True, linecolor = 'black', range = [3500, 400], nticks = 30, tickfont = dict(size = 20), linewidth = 3)
figure.update_yaxes(title = dict(text = 'Transmittance (%)', font = dict(size = 20)),  ticks = None, showgrid = False, showline = True, showticklabels = False,  linecolor = 'black', tickfont = None, linewidth = 3)
figure.update_layout(height = 600, width = 1000, showlegend = False, title = dict(text = 'FT-IR Spectra', font = dict(size = 20)))
figure.show()
```
