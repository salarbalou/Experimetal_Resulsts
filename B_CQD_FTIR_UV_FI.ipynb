```python
import numpy as np
import plotly.io as pio
from plotly.subplots import make_subplots
pio.templates.default = 'plotly_white'
import requests as req
import re
```

```python
!wget 'https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/FIEM_BCQDs.txt'
```

```python
data = np.loadtxt("https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/FIEM_BCQDs.txt")
wl = np.array(data[:, 0]) #wavelenght #nm
adin = np.array(data[:, 1]) #absorption intensity #a.u
```

```python
fig = make_subplots()
fig.add_scatter(x = wl, y = adin, mode = 'markers+lines', marker_color = 'red')
fig.update_layout(height = 800, width = 1000, title = 'BCQDs_Fluorescent Emission 448nm', title_font = dict(size = 24))
fig.update_xaxes(title = 'Wavelenght (nm)', title_font = dict(size = 25), tickfont = dict(size = 20), nticks = 20)
fig.update_yaxes(title = 'Emission Intensity (a.u) ', title_font = dict(size = 25), tickfont = dict(size = 20))
```
