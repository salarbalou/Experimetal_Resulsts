```python
import numpy as np
import requests as req
import regex as re
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
from scipy import signal
from scipy.signal import find_peaks
pio.templates.default = 'plotly_white'
```

```python
#dataB = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/XRD-BCQD.txt')
dataC = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/XRD-YCQD.txt')
dataG = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/XRD-OCQD.txt')
dataY = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/XRD-YCQD.txt')
dataR = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/XRD-RCQD.txt')

waveB, intentB  = np.arange(10, 80, step = 1e-2), np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/GCQDS.txt', delimiter = ',').reshape((7004, ))
waveC, intentC = dataC[:, 0], dataC[:, 1]
waveG, intentG = dataG[:, 0], dataG[:, 1]
waveY, intentY = dataY[:, 0], dataY[:, 1]
waveR, intentR = dataR[:, 0], dataR[:, 1]

#peaks = np.unique(np.array(wave[np.argmax(intent)], dtype = 'float64'))
#noise = np.random.normal(loc = 1, scale = 10**(-n), size = 7001)
r = 1000

figure = make_subplots()

figure.add_trace(go.Scatter(x = waveB, y = intentB+4.2*r, mode= 'lines', line_color = 'blue', name = 'XRD Spectra'))
figure.add_trace(go.Scatter(x = waveC, y = intentC+3*r, mode= 'lines', line_color = 'cyan', name = 'XRD Spectra'))
figure.add_trace(go.Scatter(x = waveG, y = intentG+2*r, mode= 'lines', line_color = 'green', name = 'XRD Spectra'))
figure.add_trace(go.Scatter(x = waveY, y = intentY+r, mode= 'lines', line_color = 'orange', name = 'XRD Spectra'))
figure.add_trace(go.Scatter(x = waveR, y = intentR, mode= 'lines', line_color = 'red', name = 'XRD Spectra'))

figure.update_xaxes(title = dict(text = '2Theta', font = dict(size = 20)), ticks = 'outside', showgrid = False, showline = True, linecolor = 'black', range = [10, 80], nticks = 30, tickfont = dict(size = 16))
figure.update_yaxes(title = dict(text = 'Intensity (a.u.)', font = dict(size = 20)),  ticks = None, showgrid = False, showline = True, showticklabels = False,  linecolor = 'black', tickfont = None)
figure.update_layout(height = 800, width = 1200, showlegend = False, title = dict(text = 'XRD Spectra', font = dict(size = 20)))
figure.show()
```
