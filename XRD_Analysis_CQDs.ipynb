```python
import numpy as np
import requests as req
import regex as re
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
from scipy import signal
from scipy.signal import find_peaks
pio.templates.default = 'plotly_white'
```

```python
data = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/XRD-OCQD.txt')
wave, intent = data[:, 0], data[:, 1]
figure = make_subplots()
figure.add_trace(go.Scatter(x = wave, y = intent, mode= 'lines', line_color = 'black', name = 'XRD Spectra'))
figure.update_xaxes(title = '2Theta', ticks = 'outside', showgrid = False, showline = True, linecolor = 'black', nticks = 50)
figure.update_yaxes(title = 'Intensity (Counts)',  ticks = 'outside', showgrid = False, showline = True, linecolor = 'black')
```
