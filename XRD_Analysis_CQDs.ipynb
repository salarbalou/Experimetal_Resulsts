```python
import numpy as np
import requests as req
import regex as re
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
from scipy import signal
from scipy.signal import find_peaks
pio.templates.default = 'plotly_white'
```

```python
n =0
```

```python
#dataB = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/XRD-BCQD.txt')
dataC = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/XRD-CCQD.txt')
dataG = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Photothermal_Study_CQDs/main/XRD-OCQD.txt')
dataY = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/XRD-YCQD.txt')
dataR = np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/XRD-RCQD.txt')

waveB, intentB  = dataC[:, 0], np.loadtxt('https://raw.githubusercontent.com/salarbalou/Experimetal_Resulsts/main/GCQDS.txt', delimiter = ',').reshape((7004, ))
waveC, intentC = dataC[:, 0], dataC[:, 1]
waveG, intentG = dataG[:, 0], dataG[:, 1]
waveY, intentY = dataY[:, 0], dataY[:, 1]
waveR, intentR = dataR[:, 0], dataR[:, 1]

#peaks = np.unique(np.array(wave[np.argmax(intent)], dtype = 'float64'))
#noise = np.random.normal(loc = 1, scale = 10**(-n), size = 7001)
r = 1500

L = 1.5406 #Wavelenght of incident X-ray
n = 1 #order of diffraction

#Peaks are manually selected
peaksB = np.array([25.64, 44.66])
peaksBI = np.squeeze(np.array([intentB[np.where(waveB == peaksB[i])[0]] for i in range(peaksB.size)])) #Corresponding intenstiy of the peaks

peaksC = np.array([26.68, 38.34, 44.68])
peaksCI = np.squeeze(np.array([intentC[np.where(waveC == peaksC[i])[0]] for i in range(peaksC.size)]))

peaksG = np.array([19.61, 38.41, 44.68])
peaksGI = np.squeeze(np.array([intentG[np.where(waveG == peaksG[i])[0]] for i in range(peaksG.size)]))

peaksY = np.array([19.61, 26.21, 30.93, 38.42, 44.68])
peaksYI = np.squeeze(np.array([intentY[np.where(waveY == peaksY[i])[0]] for i in range(peaksY.size)]))

peaksR = np.array([21.69, 38.44, 44.69])
peaksRI = np.squeeze(np.array([intentR[np.where(waveR == peaksR[i])[0]] for i in range(peaksR.size)]))

m1 = np.max(intentY)/np.max(intentB)
m2 = np.max(intentY)/np.max(intentC)

dB = np.abs((n*L)/(2*np.sin(peaksB))) #dspacing cacualated using the Brags Equaiton
dC = np.abs((n*L)/(2*np.sin(peaksC)))
dG = np.abs((n*L)/(2*np.sin(peaksG)))
dY = np.abs((n*L)/(2*np.sin(peaksY)))
dR = np.abs((n*L)/(2*np.sin(peaksR)))

figure = make_subplots()

figure.add_trace(go.Scatter(x = waveB, y = (intentB+1100)*m, mode= 'lines', line_color = 'blue', name = 'XRD Spectra')) #main plot
figure.add_trace(go.Scatter(x = np.array([10, 80]), y = np.array([4*r, 4*r]), line= dict(color = 'black', dash = 'dash'))) #baseline
figure.add_trace(go.Scatter(x = peaksB, y = peaksBI+6800, mode = 'markers+text' , text = np.round(dB, 3), textposition = 'top center' , marker = dict(color = 'black', size = 6))) #marker showing peaks
figure.add_trace(go.Scatter(x = peaksB, y = peaksBI+6800, mode = 'markers+text' , text = f'    Å', textposition = 'top right',marker = dict(color = 'black', size = 6))) #Angstrom unit position

figure.add_trace(go.Scatter(x = waveC, y = (intentC)*m2+4400, mode= 'lines', line_color = 'cyan', name = 'XRD Spectra'))
figure.add_trace(go.Scatter(x = np.array([10, 80]), y = np.array([3*r, 3*r]), line= dict(color = 'black', dash = 'dash')))
figure.add_trace(go.Scatter(x = peaksC, y = peaksCI+5200, mode = 'markers+text' , text = np.round(dC, 3), textposition = 'top center', marker = dict(color = 'black', size = 6)))
figure.add_trace(go.Scatter(x = peaksC, y = peaksCI+5200, mode = 'markers+text' , text = f'    Å', textposition = 'top right',marker = dict(color = 'black', size = 6)))

figure.add_trace(go.Scatter(x = waveG, y = intentG+2*r, mode= 'lines', line_color = 'green', name = 'XRD Spectra'))
figure.add_trace(go.Scatter(x = np.array([10, 80]), y = np.array([2*r, 2*r]), line= dict(color = 'black', dash = 'dash')))
figure.add_trace(go.Scatter(x = peaksG, y = peaksGI+2*r, mode = 'markers+text' ,text = np.round(dG, 3), textposition = 'top center', marker = dict(color = 'black', size = 6)))
figure.add_trace(go.Scatter(x = peaksG, y = peaksGI+2*r, mode = 'markers+text' , text = f'    Å', textposition = 'top right',marker = dict(color = 'black', size = 6)))

figure.add_trace(go.Scatter(x = waveY, y = intentY+r, mode= 'lines', line_color = 'orange', name = 'XRD Spectra'))
figure.add_trace(go.Scatter(x = np.array([10, 80]), y = np.array([r, r]), line= dict(color = 'black', dash = 'dash')))
figure.add_trace(go.Scatter(x = peaksY, y = peaksYI+r, mode = 'markers+text' ,text = np.round(dY, 3), textposition = 'top center', marker = dict(color = 'black', size = 6)))
figure.add_trace(go.Scatter(x = peaksY, y = peaksYI+r, mode = 'markers+text' , text = f'    Å', textposition = 'top right',marker = dict(color = 'black', size = 6)))

figure.add_trace(go.Scatter(x = waveR, y = intentR, mode= 'lines', line_color = 'red', name = 'XRD Spectra'))
figure.add_trace(go.Scatter(x = np.array([10, 80]), y = np.array([0, 0]), line= dict(color = 'black', dash = 'dash')))
figure.add_trace(go.Scatter(x = peaksR, y = peaksRI, mode = 'markers+text' ,text = np.round(dR, 3), textposition = 'top center', marker = dict(color = 'black', size = 6)))
figure.add_trace(go.Scatter(x = peaksR, y = peaksRI, mode = 'markers+text' ,text = f'    Å', textposition = 'top right', marker = dict(color = 'black', size = 6)))

figure.update_xaxes(title = dict(text = '2Theta', font = dict(size = 20)), ticks = 'outside', showgrid = False, showline = True, linecolor = 'black', range = [10, 80], nticks = 30, tickfont = dict(size = 16),linewidth = 3)
figure.update_yaxes(title = dict(text = 'Intensity (a.u.)', font = dict(size = 20)),  ticks = None, showgrid = False, showline = True, showticklabels = False,  linecolor = 'black', tickfont = None, linewidth = 3)
figure.update_layout(height = 800, width = 1000, showlegend = False, title = dict(text = 'XRD Spectra', font = dict(size = 20)))
figure.show()
```
